```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
students = {
    "Student": ["Arun", "Priya", "Karthik", "Divya", "Rahul"],
    "Maths": [85, 92, 45, 70, 35],
    "Science": [78, 88, 52, 75, 40],
    "English": [90, 95, 48, 68, 30]
}

df = pd.DataFrame(students)

print(df)
```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"]

print(df)
```

```python
df["Average"] = df["Total"] / 3

print(df)
```

```python
def grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "F"

df["Grade"] = df["Average"].apply(grade)

print(df)
```

```python
df["Result"] = np.where(df["Average"] >= 50,
                        "Pass",
                        "Fail")

print(df)
```

```python
topper = df.loc[df["Average"].idxmax()]

print(topper)
```

```python
print("Maths Average:", df["Maths"].mean())
print("Science Average:", df["Science"].mean())
print("English Average:", df["English"].mean())
```

```python
subjects = ["Maths", "Science", "English"]

averages = [
    df["Maths"].mean(),
    df["Science"].mean(),
    df["English"].mean()
]

plt.bar(subjects, averages)

plt.title("Subject Wise Performance")
plt.show()
```

```python
df["Grade"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    "Student": ["Arun", "Priya", "Karthik", "Divya", "Rahul"],
    "Maths": [85, 92, 45, 70, 35],
    "Science": [78, 88, 52, 75, 40],
    "English": [90, 95, 48, 68, 30]
}
df = pd.DataFrame(data)
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3
df["Result"] = np.where(df["Average"] >= 50, "Pass", "Fail")
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
df["Result"].value_counts().plot(kind="bar")
plt.title("Pass vs Fail")
plt.subplot(2,2,2)
subject_avg = df[["Maths","Science","English"]].mean()
subject_avg.plot(kind="bar")
plt.title("Subject Wise Average")
plt.subplot(2,2,3)
df.set_index("Student")["Average"].plot(kind="bar")
plt.title("Student Average Marks")
plt.subplot(2,2,4)
df["Average"].plot(kind="line", marker="o")
plt.title("Performance Trend")
plt.tight_layout()
plt.show()
```

```python
df["Class"] = ["CSE-A", "CSE-A", "CSE-B", "CSE-B", "CSE-A"]
class_avg = df.groupby("Class")["Average"].mean()

class_avg.plot(kind="bar")
plt.title("Class Wise Performance")
plt.show()
```
